```python
from langchain.agents import initialize_agent, AgentType, Tool
from langchain.chat_models import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI()

def calculator(math_problem):
    """
    Tool: Calculator - Evaluates mathematical expressions
    
    Args:
        math_problem (str): A mathematical expression like "5+3" or "10*7"
        
    Returns:
        str: The result of the calculation or an error message
    """
    result = eval(math_problem)
    return f"Answer: {result}"

def get_weather(city):
    """
    Tool: Weather - Returns weather information for major cities
    
    Args:
        city (str): Name of the city to get weather for
        
    Returns:
        int: The current
    """
    # Simulated weather database (in real app, this would call a weather API)
    weather_data = {
        "new york": 72,
        "london": 65, 
        "tokyo": 87
    }
    
    city = city.lower()  # Convert to lowercase for case-insensitive matching
    
    if city in weather_data:
        return f"Weather in {city}: {weather_data[city]}Â°F"
    else:
        return "City not found"
    
tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Useful for simple math like '2 + 2'"
    ),
    Tool(
        name="Weather",
        func=get_weather,
        description="Useful for getting weather information for major cities"
    )
]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

result = agent.run("What is the difference in temperature between London and Tokyo?")
print(result)
```
